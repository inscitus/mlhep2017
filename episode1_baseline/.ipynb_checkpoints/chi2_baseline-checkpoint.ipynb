```python
train_df = df[df.signal == True].values[:, 6:8]
print(len(train_df))
```

```python
train_df = np.concatenate([train_df, df[df.signal == False].values[:50000, 6:8]])
```

```python
df_train = pd.read_csv('/data_mlhep2017/episode1/DS_1_train.csv', index_col=0)
df_test = pd.read_csv('/data_mlhep2017/episode1/DS_1_test.csv', index_col=0)
del df_train['event_id']
X_train_pre = df_train.iloc[:,:-1]
X_test_pre = df_test
y_train = df[['signal']]


# -- Split for testing on the train set
#from sklearn.model_selection import train_test_split
#X = X_train_pre
#y = y_train
#X_train_pre, X_test_pre, y_train, y_test = train_test_split(X, y, random_state=0)

# -- Min/max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train_pre)
X_test = scaler.transform(X_test_pre)

# -- xgboost
import xgboost
model = xgboost.XGBClassifier(n_estimators=1000)
model.fit(X_train,y_train)
predicted = model.predict_proba(X_test)[:,1]

# -- save prediction to csv
ans = pd.DataFrame(predicted, columns=['Prediction'])
ans.index.name = 'Id'
ans.to_csv('xgb_prediction.csv', header=True)

# -- ROC AUC
#from sklearn.metrics import roc_curve, auc, roc_auc_score
#fpr, tpr, _ = roc_curve(y_test, predicted)
#roc_auc = auc(fpr, tpr)
#print('xgboost ROC AUC ' + str(roc_auc))
```
